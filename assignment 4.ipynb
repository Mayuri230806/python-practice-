*ques 12*

Create a dataset of students' marks using a NumPy array and find: 
● Highest marks 
● Lowest marks 
● Average marks 
● Students scoring above average 

```python
marks = np.array([75, 82, 90, 65, 88, 70, 95])

highest = np.max(marks)
lowest = np.min(marks)
average = np.mean(marks)

above_average = marks[marks > average]

print(highest)
print(lowest)
print( average)
print(above_average)
```

```python
arr = np.arange(1, 37)
arr
```

```python
arr=arr.reshape(6, 6)
arr
```
