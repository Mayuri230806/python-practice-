Series:
1)One-dimensional data structure	
2)Contains a single column of data	
3)Similar to a list with index	
4)Created using pd.Series()

```python
s= pd.Series([10, 20, 30, 40])
s
```

*DataFrame:
1.Two-dimensional data structure
2.Contains multiple rows and columns
3.Similar to an Excel sheet or SQL table
4.Created using pd.DataFrame()*

```python
data = {
    "Name": ["Amit", "Neha"],
    "Marks": [85, 90]
}

df = pd.DataFrame(data)
df
```

*loc[]:
Label (name) based indexing	
Uses row/column names	
End index is included	
Syntax: df.loc[row_label, column_label]	*

```python
df.loc[1, "Name"]
```

iloc[]:
Integer position based indexing
Uses row/column numbers
End index is excluded in slicing
Syntax: df.iloc[row_index, column_index]

```python
df.iloc[1, 0]
```
