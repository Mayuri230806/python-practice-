```python
word=input("enter a word:")  #helloram!

vowel_count=0
consonant_count = 0
digit_count = 0
special_count = 0

for ch in word:
     if ch in "aeiou":
        vowel_count+=1
     elif ch.isalpha():
        consonant_count+=1
     elif ch.isdigit():
         digit_count+=1
     else:
         special_count+=1

print("total vowel", vowel_count)
print("total consonant",consonant_count)
print("total digit",digit_count)
print("total special character",special_count)
```

```python
num = int(input("Enter a number: "))  #193

original = num
digits = len(str(num))
sum = 0

while num > 0:
    digit = num % 10
    sum += digit ** digits
    num = num // 10

if sum == original:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
a = int(input("Enter first side: "))
b = int(input("Enter second side: "))
c = int(input("Enter third side: "))

if a == b == c:
    print("Equilateral Triangle")

elif a == b or b == c or a == c:
    print("Isosceles Triangle")

else:
    print("Scalene Triangle")
```

```python
sentence = input("Enter a sentence: ")

freq = {}

for word in sentence.split():
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

print("Word Frequencies:")
for key, value in freq.items():
    print(key, ":", value)
```

```python
val = (1, 2, 3, 2, 4, 1, 5, 2, 3)

freq = {}

for i in val:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

print("Duplicate Elements and Their Frequency:")

for key, value in freq.items():
    if value > 1:
        print(key, ":", value)
```

```python
numbers = []

for i in range(10):
    num = int(input("Enter number: "))
    numbers.append(num)

largest = numbers[0]
smallest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num
    if num < smallest:
        smallest = num

second_largest = float('-inf')

for num in numbers:
    if num > second_largest and num != largest:
        second_largest = num

print("Largest =", largest)
print("Smallest =", smallest)
print("Second Largest =", second_largest)
```

```python
products = {}

while True:
    print("\n1 Add  2 View  3 Update  4 Delete  5 Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        name = input("Enter product name: ")
        price = input("Enter price: ")
        products[name] = price

    elif choice == 2:
        print(products)

    elif choice == 3:
        name = input("Enter product name: ")
        if name in products:
            price = input("Enter new price: ")
            products[name] = price
        else:
            print("Product not found")

    elif choice == 4:
        name = input("Enter product name: ")
        if name in products:
            del products[name]
        else:
            print("Product not found")

    elif choice == 5:
        break

    else:
        print("Invalid choice")
```

```python
students = {}

while True:
    print("\n===== STUDENT MANAGEMENT SYSTEM =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Update Marks")
    print("4. Delete Student")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    # 1. Add Student
    if choice == 1:
        name = input("Enter student name: ")
        marks = int(input("Enter marks: "))
        students[name] = marks
        print("Student added successfully!")

    # 2. View Students
    elif choice == 2:
        if len(students) == 0:
            print("No students found!")
        else:
            print("\n--- Student List ---")
            for name, marks in students.items():
                print(name, ":", marks)

    # 3. Update Marks
    elif choice == 3:
        name = input("Enter student name to update: ")
        if name in students:
            marks = int(input("Enter new marks: "))
            students[name] = marks
            print("Marks updated successfully!")
        else:
            print("Student not found!")

    # 4. Delete Student
    elif choice == 4:
        name = input("Enter student name to delete: ")
        if name in students:
            del students[name]
            print("Student deleted successfully!")
        else:
            print("Student not found!")

    # 5. Exit
    elif choice == 5:
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Try again.")
```
